1. Implementar método de Euler e Método Runge-kutta de quarta ordem para resolver um PVI composto para equação diferencial de primeira ordem.

A implementação do item 1 está no arquivo pvi_q1.py. Importaremos o arquivo e mostraremos um exemplo.

```python
import pvi_q1 as q1

# Instanciando um exemplo

ponto_inicial = (0,2)
passo = 0.01
intervalo = (0,0.5)
funcao = "-2 * (x**2) * (y**2)"

ex1 = q1.pvi(ponto_inicial, passo, intervalo, funcao)
```

```python
# Resolvendo o exemplo com o método de Euler
ex1.euler()

# Mostrando os 5 primeiros pontos
for i in range(5):
    print(ex1.pontos[i])

# Exibindo o gráfico
ex1.plottar()
```

```python
# Resolvendo o exemplo com o método de Runge Kutta de Quarta Ordem
ex1.rg4()

# Mostrando os 5 primeiros pontos
for i in range(5):
    print(ex1.pontos[i])

# Exibindo o gráfico
ex1.plottar()
```
